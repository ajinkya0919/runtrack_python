```python
import re
from collections import Counter

# Find words repeated only once in a text document
def find_unique_words(text):
    # Use for word tokenization
    words = re.findall(r'\b\w+\b', text.lower())
    
    word_counts = Counter(words)
    
    unique_words = [word for word, count in word_counts.items() if count == 1]
    
    return unique_words

with open("data.txt", "r") as file:
    document_text = file.read()

# Find unique words in the document
unique_words = find_unique_words(document_text)
print(unique_words)

```

```python
# List of all pokemons
df= pd.read_csv("pokemon.csv", usecols= ["Name"])
df['Name'] = df['Name'].str.lower()

df_name=df.values.tolist()

# Flattening the sublists into a single array of pokemons
df_name = [item for sublist in df_name for item in sublist]
print(df_name)
```
