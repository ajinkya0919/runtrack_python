```python
f = open("data.txt", "r")
text = f.read()
joined = ''.join(letter for letter in text if letter.isalpha())

joined = joined.lower()
# print(len(joined))
```

```python
chars = set(joined)

res = {}
for char in joined:
    res[char] = res.get(char, 0) + 1

total = 0
for i in res.keys():
    total += res[i] 

for key in res.keys():
    res[key] = res[key]*100/total
print(res)
```

```python
import matplotlib.pyplot as plt
plt.bar(res.keys(), res.values())
# plt.show()
```
