```python
import matplotlib.pyplot as plt
```

```python
f = open("data.txt", "r")
text = f.read()
words = text.split()

freq = {}

for word in words:
    if len(word) in freq:
        freq[len(word)] += 1
    else:
        freq[len(word)] = 1

freq = dict(sorted(freq.items()))
print(freq)
```

```python
total = 0
for i in freq.keys():
    total += freq[i]

for i in freq.keys():
    freq[i] = freq[i]*100/total
```

```python
plt.xticks([i for i in range(20)])
plt.bar(freq.keys(), freq.values())
```
