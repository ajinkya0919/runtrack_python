```python
import re 
import matplotlib.pyplot as plt
import nltk
```

```python
with open('data.txt', 'r') as file:
    text = file.read()


period = re.split(r'[.!?]', text)
total_words = []

length_map = {}
i = 0
for line in period:
    words = re.findall(r'\b\w+\b', line)
    total_words.append(words)
    length_map[i] = len(words)
    i += 1

print(length_map)
```

```python
plt.figure(figsize=[12,8])
plt.hist(x = length_map.values(),bins = 60)
plt.plot()
plt.show()
#plt
```

```python
from fuzzywuzzy import fuzz

# Define a list of words
word_list = re.findall(r'\b\w+\b', text)

# Function to generate similar sounding words
# The threshold sets limits for how flexible the output needs to be
def generate_similar_sounding_words(input_word, word_list, threshold=70):
    similar_words = []
    for word in word_list:
        similarity = fuzz.ratio(input_word, word)
        if similarity >= threshold:
            similar_words.append(word)

    return similar_words

input_word = input("Enter a word or phrase: ") # Lorem ipsum for example

# Generate and print similar sounding words
similar_words = generate_similar_sounding_words(input_word, set(word_list))

if similar_words:
    print(f"Similar sounding words to '{input_word}': {', '.join(similar_words)}")
else:
    print(f"No similar sounding words found for '{input_word}'.")
```
